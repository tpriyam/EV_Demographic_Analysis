```python
def createList(field, user):
    ls_field = [0]*len(user)
    for i in range(len(user)):
        ls_field[i] = user.iloc[i][field]
    return ls_field
## Pass ls_field as the exact field name in the dataframe
```

```python
id = createList("id",user)
name = createList("name",user)
screen_name = createList("username",user)
desc = createList("description",user)
lang = createList("lang",df)
img = createList("profile_image_url",user)
```

```python
import json
```

```python
with open('sample_data.json', 'w') as outfile:
    for i in range(len(user)):
        values = {"id" : id[i], "name" : name[i],"screen_name" : screen_name[i] ,"description" : desc[i] ,"lang" : lang[i],"img_path" : img[i]}
        json.dump(values, outfile, indent = None)
        outfile.write("\n")
```
