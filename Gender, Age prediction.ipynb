```python
import time
```

```python
prev = time.time()

for i in range(len(dic)):
    tmp = dic[ds.iloc[i]['user']['id']]['gender']
    max_gender = max(tmp, key = tmp.get)
    ds.loc[i,"gender"] = max_gender
#     print(max_gender)
    a = dic[ds.iloc[i]['user']['id']]['age']
    max_key = max(a, key = a.get)
    ds.loc[i,"age"] = max_key
#     print(max_key)
    if i%1000==0:
        print(i," complete")
        print(time.time()-prev)
        
elapsed_time_fl = (time.time() - prev)
print("Final time: ",elapsed_time_fl)
```
