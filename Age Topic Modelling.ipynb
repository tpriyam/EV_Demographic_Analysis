```python
import re
import nltk
from nltk.corpus import stopwords
import string
import pandas as pd
```

```python
nltk.download('stopwords')
```

```python
def clean_text(text):
    url_pattern = re.compile(r'https?://\S+|www\.\S+')
    text = url_pattern.sub(r'', text)
    text = re.sub('\S*@\S*\s?', '', text)
    text = re.sub('\s+', ' ', text)
    text = re.sub("\'", "", text)
    tokens = text.split()
    table = str.maketrans('', '', string.punctuation)
    tokens = [w.translate(table) for w in tokens]
    tokens = [word for word in tokens if word.isalpha()]
    stop_words = set(stopwords.words('english'))
    tokens = [w for w in tokens if not w in stop_words]
    tokens = [word for word in tokens if len(word) > 1]
    
    return tokens
```

## <=18 TOPICS

```python
df1['cleanText'] = df1['text'].apply(lambda x: re.split('https:\/\/.*', str(x))[0])
```

```python
df1['cleanText'] = \
df1['cleanText'].map(lambda x: re.sub('[,\.!?]', '', x))
# Convert the titles to lowercase
df1['cleanText'] = \
df1['cleanText'].map(lambda x: x.lower())
# Print out the first rows of papers
df1['cleanText'].head()
```

```python
from nltk.corpus import stopwords
stop_words = stopwords.words('english')
stop_words.extend(['from', 'subject', 're', 'edu', 'use','electric','vehicle','vehicles','car','cars','rt','ev'])

def sent_to_words(sentences):
    for sentence in sentences:
        # deacc=True removes punctuations
        yield(gensim.utils.simple_preprocess(str(sentence), deacc=True))
        
def remove_stopwords(texts):
    return [[word for word in simple_preprocess(str(doc)) 
             if word not in stop_words] for doc in texts]
## add electric, vehicle, rts etc in stopwords
```

```python
data1 = df1['cleanText'].values.tolist()
data_words1 = list(sent_to_words(data1))
# remove stop words
data_words1 = remove_stopwords(data_words1)
print(data_words1[:1][0][:30])
```

```python
import gensim.corpora as corpora
# Create Dictionary
id2word1 = corpora.Dictionary(data_words1)
# Create Corpus
texts1 = data_words1
# Term Document Frequency
corpus1 = [id2word1.doc2bow(text) for text in texts1]
# View
print(corpus1[:1][0][:30])
```

```python
import os
import pyLDAvis.gensim_models
```

```python
from pprint import pprint
# number of topics
num_topics = 8
# Build LDA model
lda_model_1 = gensim.models.LdaMulticore(corpus=corpus1,
                                       id2word=id2word1,
                                       num_topics=num_topics)
# Print the Keyword in the 10 topics
pprint(lda_model_1.print_topics())
doc_lda1 = lda_model_1[corpus1]
```

```python
import pyLDAvis.gensim_models
import pickle 
import pyLDAvis
# Visualize the topics
pyLDAvis.enable_notebook()
LDAvis_data_filepath = os.path.join('fol1'+str(num_topics))
# # this is a bit time consuming - make the if statement True
# # if you want to execute visualization prep yourself
if 1 == 1:
    LDAvis_prepared1 = pyLDAvis.gensim_models.prepare(lda_model_1, corpus1, id2word1)
    with open(LDAvis_data_filepath, 'wb') as f:
        pickle.dump(LDAvis_prepared1, f)
# load the pre-prepared pyLDAvis data from disk
with open(LDAvis_data_filepath, 'rb') as f:
    LDAvis_prepared1 = pickle.load(f)
pyLDAvis.save_html(LDAvis_prepared1, 'fol1'+ str(num_topics) +'.html')
LDAvis_prepared1
```

```python
import matplotlib.pyplot as plt
```

```python
from wordcloud import WordCloud
def create_wordcloud(model, topic):
    text = {word: value for word, value in model.show_topic(topic)}
    wc = WordCloud(background_color="white", max_words=1000)
    wc.generate_from_frequencies(text)
    plt.imshow(wc, interpolation="bilinear")
    plt.axis("off")
    plt.title("Topic" + " "+ str(topic))
    plt.show()
```

```python
for i in range(0,num_topics):
    create_wordcloud(lda_model_1, topic=i)
```
